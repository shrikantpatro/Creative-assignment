```python
import urllib.request
import os
import pytesseract
import random
from PIL import Image
pytesseract.pytesseract.tesseract_cmd="C:\\Users\\Shridhar Palo\\AppData\\Local\\Tesseract-OCR\\tesseract.exe"
text_list =[]
empty_text = ''
#Download file and egxtract text from image
for i in range(len(url_arr)):
    try:
        resource = urllib.request.urlopen(url_arr[i])
        file_name = str(os.getpid())+str(random.randint(1, 100)) +".jpg"
        output = open(file_name,"wb")
        output.write(resource.read())
    
        im=Image.open(file_name) 
        text = pytesseract.image_to_string(im,lang='eng')
        print("Begin\n")
        print("\n")
        print("{}\n".format(i))
        print("{}\n".format(text))
        print("End")
        print("\n")
        text_list.append(text)
        output.close()
        os.remove(file_name)
    except:
        text_list.append(empty_text)
print(text_list)
```

```python
#Removing the stop words from test.txt file and this will return file without stopwords
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
# Get nltk stopword list into a set.
stop_words = set(stopwords.words('english'))
# Open and read in a text file.
txt_file = open("test.txt")
txt_line = txt_file.read()
txt_words = txt_line.split()
# stopwords found counter.
sw_found = 0
sw_list =[]
# If each word checked is not in stopwords list,
# then append word to a new text file.
for check_word in txt_words:
    if not check_word.lower() in stop_words:
        # Not found on stopword list, so append.
        appendFile = open('stopwords-removed.txt','a')
        appendFile.write(" "+check_word)
        appendFile.close()
    else:
        # It's on the stopword list
        sw_found +=1
        sw_list.append(check_word)
print(sw_found,"stop words found and removed")
print("Saved as 'stopwords-removed.txt' ")
```

```python
#Count frequency of each word from stopwords-removed.txt file
import re
import string
frequency = {}
document_text = open('stopwords-removed.txt', 'r')
text_string = document_text.read().lower()
match_pattern = re.findall(r'\b[a-z]{3,15}\b', text_string)
 
for word in match_pattern:
    count = frequency.get(word,0)
    frequency[word] = count + 1
     
frequency_list = frequency.keys()
 
for words in frequency_list:
    print(words, frequency[words])
```

```python
#Sorting the wird frequency dectionary in ascending order so that required most frequently appeared word should appear toward end
from operator import itemgetter
from collections import OrderedDict
sorted_x = OrderedDict(sorted(frequency.items(), key=itemgetter(1)))
```
